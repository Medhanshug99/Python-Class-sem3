```python
def is_prime(num):
  if num <= 1:
    return False
  for i in range (2, num):
    if num % i == 0:
      return False
  return True

num1_str = input("Enter the number : ")
try:
  num1_int = int(num1_str)
  if is_prime(num1_int) :
    print(f"{num1_str} is a prime number")
  else:
    print(f"{num1_str} is not a prime number")
except ValueError:
  print(f"Invalid input: '{num1_str}' is not a valid integer. Please enter a whole number.")
```

Armstrong Number - Sum of the numbers raised to power of its length

```python
def is_armstrong_num(n_str):
  num_int = int(n_str)
  length = len(n_str)
  sum_of_powers = 0
  temp_num = num_int

  while temp_num > 0:
    digit = temp_num % 10
    sum_of_powers += digit ** length
    temp_num //= 10
  return sum_of_powers == num_int

a = input("Enter the number : ")
try:
  if is_armstrong_num(a):
    print(f"{a} is an Armstrong number")
  else:
    print(f"{a} is not an Armstrong number")
except ValueError:
  print(f"Invalid input: '{a}' is not a valid integer. Please enter a valid number.")
```

Reverse the number

```python
def reverse_num(n_str):
  num_int = int(n_str)
  reversed_number = 0
  temp = num_int

  while temp > 0:
    digit = temp % 10
    reversed_number = reversed_number * 10 + digit
    temp //= 10
  return reversed_number

s_str = input("Enter the number : ")
try:
  reversed_result = reverse_num(s_str)
  print(f"The reversed number is: {reversed_result}")
except ValueError:
  print(f"Invalid input: '{s_str}' is not a valid integer. Please enter a whole number.")
```

FActorial

```python
def factorial(n):
  fact = 1
  for i in range(1,n+1):
    fact*=i
  return fact

print(factorial(5))
```

```python
def fact(n):
  if n == 0 or n == 1:
    return 1
  return n * fact(n-1)
print(fact(6))
```
